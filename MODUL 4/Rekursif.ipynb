# Definisi :
## 1. Rekursif
Rekursif adalah teknik pemrograman atau pendekatan algoritma di mana sebuah fungsi memanggil dirinya sendiri secara berulang untuk menyelesaikan suatu masalah. Teknik ini bekerja dengan cara memecah masalah yang kompleks menjadi sub-masalah yang lebih kecil dan serupa hingga mencapai kondisi paling sederhana yang dapat langsung diselesaikan.

### Cara Kerja Umum:
* Menentukan **Base Case** (Kasus Dasar): Ini adalah kondisi berhenti atau titik akhir di mana fungsi tidak lagi memanggil dirinya sendiri, sehingga mencegah terjadinya perulangan tak terbatas (*infinite loop*).
* Menentukan **Recursive Case** (Kasus Rekursif): Bagian di mana fungsi memanggil dirinya sendiri dengan argumen atau nilai masukan yang dimodifikasi, sehingga secara bertahap kondisinya semakin mendekati *Base Case*.
* Setiap kali fungsi memanggil dirinya sendiri, status eksekusi saat itu akan ditunda dan disimpan di dalam tumpukan memori komputer (*call stack*).
* Begitu *Base Case* terpenuhi, fungsi berhenti memanggil dirinya sendiri dan mulai mengembalikan hasil eksekusi dari tumpukan memori secara mundur hingga mencapai pemanggilan fungsi yang pertama kali.


### Implementasi 1: Faktorial
* Faktorial adalah operasi matematika yang mengalikan sebuah bilangan bulat positif dengan semua bilangan bulat positif di bawahnya (contoh: $5! = 5 \times 4 \times 3 \times 2 \times 1$).
* **Base Case:** Jika nilai $n = 0$ atau $n = 1$, maka fungsi mengembalikan nilai 1.
* **Recursive Case:** Jika $n > 1$, maka fungsi mengembalikan nilai $n \times faktorial(n-1)$.
* Fungsi akan terus memanggil dirinya sendiri untuk mengalikan angka saat ini dengan angka yang lebih kecil hingga mencapai angka 1, lalu hasil akhirnya dihitung secara mundur.

# **CODE**

```python
def faktorial(n):
    # Base Case: jika n adalah 0 atau 1
    if n == 0 or n == 1:
        return 1
    # Recursive Case: n dikali faktorial dari n-1
    else:
        return n * faktorial(n-1)

# Contoh Penggunaan
angka = 5
print(f"Hasil Faktorial dari {angka}! :", faktorial(angka))
```

### Implementasi 2: Menara Hanoi (Tower of Hanoi)
* Menara Hanoi adalah permainan teka-teki matematis dengan tujuan memindahkan tumpukan cakram dari tiang asal ke tiang tujuan, dengan syarat cakram yang lebih besar **tidak boleh** diletakkan di atas cakram yang lebih kecil.
* **Base Case:** Jika hanya ada 1 cakram ($n = 1$), langsung pindahkan cakram tersebut dari tiang asal ke tiang tujuan.
* **Recursive Case:** Untuk memindahkan $n$ cakram, logika pemecahannya dibagi menjadi tiga pemanggilan berulang:
  1. Pindahkan $n-1$ cakram bagian atas dari tiang asal ke tiang bantuan (menggunakan tiang tujuan sebagai perantara).
  2. Pindahkan 1 cakram terbesar (yang tersisa di paling bawah) dari tiang asal langsung ke tiang tujuan.
  3. Pindahkan kembali $n-1$ cakram dari tiang bantuan ke tiang tujuan (menggunakan tiang asal sebagai perantara).

# **CODE**

```python
def menara_hanoi(n, asal, tujuan, bantuan):
    # Base Case: jika hanya ada 1 cakram, langsung pindahkan
    if n == 1:
        print(f"Pindahkan cakram 1 dari tiang {asal} ke tiang {tujuan}")
        return
    
    # Recursive Case:
    # 1. Pindahkan n-1 cakram dari asal ke bantuan
    menara_hanoi(n-1, asal, bantuan, tujuan)
    
    # 2. Pindahkan cakram terbesar (n) dari asal ke tujuan
    print(f"Pindahkan cakram {n} dari tiang {asal} ke tiang {tujuan}")
    
    # 3. Pindahkan n-1 cakram dari bantuan ke tujuan
    menara_hanoi(n-1, bantuan, tujuan, asal)

# Contoh Penggunaan
jumlah_cakram = 3
print(f"Langkah-langkah memindahkan {jumlah_cakram} cakram:")
menara_hanoi(jumlah_cakram, 'A', 'C', 'B')
```
