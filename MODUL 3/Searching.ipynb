# Definisi :
## 1. Unordered List Sequential Search
Unordered List Sequential Search adalah algoritma pencarian paling dasar yang memeriksa setiap elemen dalam daftar yang **tidak terurut** secara berurutan satu per satu. Karena datanya acak, algoritma ini harus memeriksa hingga elemen terakhir jika target tidak ditemukan.

### Cara Kerja:
* Memulai pencarian dari elemen pertama (indeks ke-0) di dalam daftar.
* Membandingkan elemen yang sedang diperiksa saat ini dengan nilai target yang dicari ($target$).
* Jika elemen saat ini sama dengan target ($Ist[i] == target$), maka pencarian berhasil dan indeks dikembalikan.
* Jika tidak sama, lanjut ke elemen berikutnya.
* Proses ini terus dilakukan hingga target ditemukan atau pencarian mencapai batas akhir daftar (jika target tidak ada, semua elemen wajib diperiksa).

```python
# 1. Unordered List Sequential Search (Pencarian Tidak Terurut)
def seqSearch(listData, data):
    ind = 0
    found = False
    
    while ind < len(listData) and not found:
        if listData[ind] == data:
            found = True
        else:
            ind = ind + 1
            
    return found  # Mengembalikan True jika ada, False jika tidak ada


# 2. Ordered List Sequential Search (Pencarian Terurut)
def orderedSeqSearch(listData, data):
    ind = 0
    found = False
    stop = False
    position = []
    
    while ind < len(listData) and not found and not stop:
        if listData[ind] == data:
            found = True
            position.append(ind)
        else:
            # Berhenti lebih awal jika elemen saat ini lebih besar dari data yang dicari
            if listData[ind] > data:
                stop = True
            else:
                ind = ind + 1
                
    if found:
        return ind  # Mengembalikan indeks dari data yang ditemukan
    else:
        return 'Data tidak ada'


# ==========================================
# CONTOH PENGGUNAAN
# ==========================================
if __name__ == "__main__":
    # Pengujian Unordered Search
    data_acak = [11, 23, 58, 31, 56, 77, 43, 12, 65, 19]
    print("--- seqSearch (Unordered) ---")
    print("Apakah 31 ada di data acak? :", seqSearch(data_acak, 31))
    print("Apakah 99 ada di data acak? :", seqSearch(data_acak, 99))
    
    # Pengujian Ordered Search
    data_terurut = [11, 12, 19, 23, 31, 43, 56, 58, 65, 77]
    print("\n--- orderedSeqSearch (Ordered) ---")
    print("Mencari 56 di data terurut  : Indeks ke-", orderedSeqSearch(data_terurut, 56))
    print("Mencari 50 di data terurut  :", orderedSeqSearch(data_terurut, 50))
```

# Definisi :
## 2. Ordered List Sequential Search
Ordered List Sequential Search adalah algoritma pencarian sekuensial yang diterapkan pada daftar yang **sudah terurut** (misalnya, dari kecil ke besar). Keuntungannya adalah pencarian dapat dihentikan lebih awal jika algoritma menemukan elemen yang nilainya lebih besar dari target, karena elemen di sebelahnya pasti akan lebih besar lagi.

### Cara Kerja:
* Memulai pencarian dari elemen pertama (indeks ke-0) di dalam daftar.
* Membandingkan elemen yang sedang diperiksa dengan nilai target ($target$).
* Jika elemen saat ini sama dengan target ($Ist[i] == target$), pencarian berhasil dan indeks dikembalikan.
* Jika elemen saat ini **lebih besar** dari target ($Ist[i] > target$), maka pencarian langsung **dihentikan** karena tidak mungkin target berada di sisa daftar tersebut.
* Jika elemen saat ini masih lebih kecil dari target, lanjut ke elemen berikutnya hingga mencapai akhir daftar.
