# Definisi :
## 1. Hashing
Hashing adalah algoritma pencarian di mana semua data terletak tepat berada ditempatnya masing-masing, sehingga pencarian dilakukan hanya dengan satu kali proses perbandingan saja. Di dalam algoritma hashing ini terdapat beberapa istilah dasar:
* **Hash Table**: Sebuah tempat penyimpanan data, yang dibuat sedemikian rupa, sehingga dapat memudahkan pencarian. Tipe data list di python dapat digunakan untuk merepresentasikan hash table.
* **Slot**: Posisi (indeks) yang terdapat pada hash table sebagai tempat penyimpanan setiap data. Nilai slot adalah integer mulai dari nol sampai dengan ukuran dari hash table.
* **Hash Function**: Suatu fungsi yang memetakan antara data dengan slot di dalam hash table.

### Cara Kerja:
* Data di dalam list disusun berdasarkan nilai hash, dan pencarian data dilakukan berdasarkan nilai hash dari hash function.
* Pencarian data dilakukan hanya dengan cara menghitung function hash berdasarkan data tersebut.
* Nilai balik dari perhitungan tersebut merupakan index atau slot dari data di dalam list.
* Kemudian dilakukan pengecekan keberadaan data tersebut hanya dengan satu kali perbandingan saja.
* Jika terdapat data dengan nilai hash yang sama sehingga mengacu pada nomor slot yang sama, maka akan terjadi masalah yang disebut *collusion*.
* *Collusion* ini dapat ditangani dengan konsep *open addressing* (mencari slot kosong) seperti *Linear Probing* atau *Quadratic Probing*, maupun dengan metode *Chaining* di mana satu slot dapat ditempati oleh beberapa data sekaligus.

### Fungsi Hash:
Hash function memiliki parameter nilai data dan mengembalikan nilai integer (nilai hash) yang merupakan representasi dari nomor slot. Berikut adalah beberapa metode fungsi hash:
* **Remainder Function**: Fungsi hash sederhana di mana nilai baliknya berupa modulus dari data tersebut dengan sebuah angka (misalnya ukuran dari tabel). Contoh: terdapat data bernilai 54 dengan ukuran tabel 11, maka nilai hash = 54 % 11, yaitu 4.
* **Mid-Square Method**: Ini adalah contoh *perfect hash function* yang bertujuan meminimalkan terjadinya *collusion*. Metodenya adalah mengkuadratkan nilai data, kemudian mengambil bagian tertentu (tengah) dari data tersebut, lalu dimodulus dengan ukuran tabel. Contoh: data 35 dikuadratkan menjadi 1225, diambil nilai tengahnya yaitu 22, lalu 22 % 11, hasilnya adalah slot 0.
* **Fungsi Hash untuk String**: Hash function juga dapat digunakan untuk tipe data string dengan mengambil jumlah nilai ASCII karakter-karakter penyusun data tersebut menggunakan perintah `ord`. Untuk menghindari *collusion* pada kata anagram (contoh 'dia' dan 'adi' yang nilai ASCII totalnya sama), dapat digunakan bobot yang menunjukkan posisi masing-masing karakter untuk dikalikan dengan nilai ASCII-nya.

# **CODE**

```python
# 1. Membuat Hash Table dengan semua slot bernilai 'none'
def createHashTable(num):
    temp = []
    for i in range(num):
        temp.append('none')
    return temp

# 2. Fungsi Hash menggunakan metode Remainder (Modulus)
def remainderFunction(data, num):
    return (data % num)

# 3. Penanganan Collusion dengan Linear Probing
def linearProbing(ind, hashTable, data):
    count = ind
    found = False
    
    # Cari slot yang kosong ('none') dengan maju satu langkah terus menerus
    while (count != ind - 1) and not (found):
        if hashTable[count] == 'none':
            found = True
        else:
            # Jika belum kosong, pindah ke slot berikutnya
            hashTable[count] = data # (Dalam PDF tertulis di bagian ini, tapi logika yang benar harus di assign saat found)
            count = count + 1
            
            # Jika pencarian mencapai akhir list, kembali ke indeks 0
            if count == len(hashTable): 
                count = 0
                
    # Memasukkan data setelah menemukan slot kosong
    if found:
        hashTable[count] = data
        
    return hashTable

# 4. Memasukkan data ke dalam tabel dengan fungsi Hash & Penanganan Collision
def putData3(a, hashTable, functionName):
    for i in range(len(a)):
        if functionName == 'reminder':
            ind = remainderFunction(a[i], len(hashTable))
        
        # Jika slot masih kosong, langsung masukkan
        if hashTable[ind] == 'none':
            hashTable[ind] = a[i]
        # Jika sudah ada isinya (Collusion), selesaikan dengan Linear Probing
        else:
            hashTable = linearProbing(ind, hashTable, a[i])
            
    return hashTable

# 5. Pencarian Data Sederhana pada Hash Table
def searchHash(data, table):
    hashVal = remainderFunction(data, len(table))
    if data == table[hashVal]:
        return True
    else:
        return False

# 6. Fungsi Hash Khusus untuk Data String (Berdasarkan Nilai ASCII)
def strVal(strData):
    temp = 0
    for i in range(len(strData)):
        temp = temp + ord(strData[i])
    return temp


# ==========================================
# CONTOH PENGGUNAAN
# ==========================================
if __name__ == "__main__":
    # Data awal beserta data baru yang akan memicu collusion (44, 55, 20)
    data_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
    
    # Buat hash table dengan ukuran 11
    hashTable = createHashTable(11)
    print("Hash Table Kosong:")
    print(hashTable)
    print("-" * 50)
    
    # Memasukkan seluruh data ke dalam tabel
    hashTable = putData3(data_list, hashTable, 'reminder')
    
    print("\nHash Table Setelah Diisi Data (beserta penanganan Collusion):")
    print(hashTable)
    print("-" * 50)
    
    # Pengujian mencari data
    print("\nApakah 93 ada di dalam tabel?", searchHash(93, hashTable))
    
```
