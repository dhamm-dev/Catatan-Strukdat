# +-**DEFINISI QUEUE**

Sebuah kumpulan item berurutan di mana penambahan item baru terjadi pada satu ujung, yang disebut “belakang” (rear), dan penghapusan item yang sudah ada terjadi pada ujung lainnya, yang umumnya disebut “depan” (front).
Prinsip pengurutan ini terkadang disebut FIFO, first-in first-out (pertama masuk, pertama keluar). Ini juga dikenal sebagai prinsip “pertama datang, pertama dilayani” (first-come first-served).

### **Oprasi dalam Queue**

* Queue(): Membuat sebuah queue (antrean) baru yang kosong. Operasi ini tidak membutuhkan parameter dan akan mengembalikan sebuah antrean kosong.

* enqueue(item): Menambahkan sebuah item baru ke bagian belakang (rear) antrean. Operasi ini membutuhkan parameter berupa item yang akan dimasukkan dan tidak mengembalikan nilai apa pun.

* dequeue(): Menghapus item terdepan (front) dari antrean. Operasi ini tidak membutuhkan parameter dan akan mengembalikan item yang dihapus tersebut. Struktur antrean akan berubah (termodifikasi).

* isEmpty(): Menguji atau memeriksa apakah antrean dalam keadaan kosong. Operasi ini tidak membutuhkan parameter dan akan mengembalikan nilai boolean (True atau False).

* size(): Mengembalikan jumlah item yang ada di dalam antrean. Operasi ini tidak membutuhkan parameter dan akan mengembalikan nilai bilangan bulat (integer).

### **Contoh beberapa code di Queue dan Dequeue dengan class**

```python
class Queue:

    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []
        
    def enqueue(self, item):
        self.items.insert(0,item)

    def dequeue(self):
        return self.items.pop()

    def size(self):
        return len(self.items)
```

### **Definisi Dequeue**

Dequeue juga dikenal sebagai double-ended queue (antrean dua ujung), adalah kumpulan item berurutan yang mirip dengan queue (antrean). Struktur ini memiliki dua ujung, yaitu depan (front) dan belakang (rear), dan item-item tersebut tetap berada pada posisinya di dalam kumpulan tersebut.

### **Oprasi dalam Dequeue**

* Deque(): Membuat sebuah deque baru yang kosong. Operasi ini tidak membutuhkan parameter dan akan mengembalikan sebuah deque kosong.

* addFront(item): Menambahkan sebuah item baru ke bagian depan (front) dari deque. Operasi ini membutuhkan parameter berupa item yang akan ditambahkan dan tidak mengembalikan nilai apa pun.

* addRear(item): Menambahkan sebuah item baru ke bagian belakang (rear) dari deque. Operasi ini membutuhkan parameter berupa item yang akan ditambahkan dan tidak mengembalikan nilai apa pun.

* removeFront(): Menghapus item terdepan dari deque. Operasi ini tidak membutuhkan parameter dan akan mengembalikan item yang dihapus tersebut. Struktur deque akan berubah (termodifikasi).

* removeRear(): Menghapus item paling belakang dari deque. Operasi ini tidak membutuhkan parameter dan akan mengembalikan item yang dihapus tersebut. Struktur deque akan berubah (termodifikasi).

* isEmpty(): Menguji atau memeriksa apakah deque dalam keadaan kosong. Operasi ini tidak membutuhkan parameter dan akan mengembalikan nilai boolean (True atau False).

* size(): Mengembalikan jumlah item yang ada di dalam deque. Operasi ini tidak membutuhkan parameter dan akan mengembalikan nilai bilangan bulat (integer).
