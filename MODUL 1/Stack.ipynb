# **Definisi Stack**

Sebuah kumpulan item berurutan di mana penambahan item baru dan penghapusan item yang sudah ada selalu terjadi pada ujung yang sama. Ujung ini umumnya disebut sebagai “puncak” (top). Ujung yang berlawanan dengan puncak dikenal sebagai “dasar” (base). Prinsip pengurutan ini terkadang disebut LIFO (Last-In, First-Out / Terakhir Masuk, Pertama Keluar) atau FILO (First-In, Last-Out / Pertama Masuk, Terakhir Keluar).

### **Prinsip LIFO (Last In First Out)**


    Prinsip LIFO berarti elemen yang paling akhir ditambahkan ke dalam stack adalah elemen pertama yang akan dihapus atau diambil.
* Elemen baru selalu ditambahkan (didorong/*pushed*) ke posisi paling atas.
* Penghapusan (*pop*) juga hanya terjadi dari posisi atas.
* Aturan ini memastikan urutan yang ketat : **terakhir masuk -> pertama keluar**

# **Operasi Basic Pada Stack**

* Stack() : membuat sebuah stack (tumpukan) baru yang kosong. Operasi ini tidak membutuhkan parameter dan mengembalikan sebuah stack yang kosong.
* Push(): Menambahkan elemen baru ke atas stack.
* Pop(): Menghapus dan mengembalikan elemen teratas dari stack.
* Peek(): Mengembalikan (melihat) elemen teratas (terakhir) pada stack.
* isEmpty(): Memeriksa apakah stack tersebut kosong.
* Size(): Mengetahui jumlah elemen di dalam stack.

### **CONTOH CODE**

```python
def stack():
    s = []
    return (s)
def push(s,data):
    s.append(data)
def pop(s):
    data = s.pop()
    return (data)
def peek(s):
    return (s[len(s)-1])
def isEmpty(s):
    return (s==[])
def size(s):
    return (len(s))
```

### **Contoh code stack dengan class**

```python
class Stack:

    def __init__(self):
        self.items = []
        
    def isEmpty(self):
        return self.items == []
        
    def push(self, item):
        self.items.append(item)
        
    def pop(self):
        return self.items.pop()
        
    def peek(self):
        return self.items[len(self.items)-1]
        
    def size(self):
        return len(self.items)
```

Struktur data direpresentasikan dengan lists.

Posisi top dari suatu stacks berada pada posisi terakhir pada list, oleh karena itu, operasi *push* pada stack, menggunakan method *append* dari list, sehingga dengan operasi 8append* ini, penambahan data baru terletak pada posisi akhir pada list.

Misalkan terdapat variabel dataStack yang menggunakan stacks berisikan data<br>
dataStack = \[4,8,1,0]<br>
maka posisi top berada pada 0. Sehingga ketika dilakukan push(dataStack, 72), maka dataStack akan berubah menjadi<br>
dataStack = \[4,8,1,72].<br>
Sedangkan operasi *pop* pada list, menggunakan method *pop* dari list, karena method *pop* ini menghapus data yang terletak pada posisi akhir dari list.

### **Contoh Penggunaan Operasi Stack**

```python
data = stack()
push(data,50)
push(data,60)
pop(data)
push(data,40)
print(data)
```

7️⃣Buatlah fungsi untuk mengecek apakah suatu string merupakan palindrome menggunakan stack.

Input: "KATAK"

Output: Palindrome

```python
def cekPalindrome(kata) :
    cek = stack()
    for huruf in kata :
        push(cek,huruf)
    wadah = ''
    while not(isEmpty(cek)) :
        wadah = wadah + pop(cek)
    if wadah==kata:
        return "Palindrome"
    else:
        return "Bukan palindrome"
```

```python
print(cekPalindrome("KATAK"))
print(cekPalindrome("JAMUR"))
```

### 1. Riwayat Undo (Simulasi Undo Text Editor)

Buatlah program sederhana untuk mensimulasikan fitur Undo pada editor teks.

Ketentuan:

User memasukkan beberapa teks.

Setiap input disimpan ke dalam stack.

Jika user mengetik UNDO, maka teks terakhir dihapus.

```python
def undo():
    teks = stack()
    while True :
        kata = input("input : ")
        if kata == "UNDO" :
            if isEmpty(teks) :
                return print("UNDO GAGAL, STACK ANDA KOSONG!")
            else :
                pop(teks)
                break
        else :
            push(teks,kata)
    return print("Output: ",peek(teks))
    
```

```python
undo()
```

### 2. Buat program untuk membalik urutan kata dalam sebuah kalimat menggunakan stack.<br>
Input: Saya belajar struktur data<br>
Output:data struktur belajar Saya

```python
def balik(kalimat):
    awal_kata = kalimat.split()
    wadah = stack()
    for kata in awal_kata:
        push(wadah, kata)
    hasil = ''
    while True:
        if isEmpty(wadah) :
            break
        else :
            hasil = hasil + pop(wadah) + " "
    return print(hasil)
```

```python
kalimat = "saya belajar struktur data"
balik(kalimat)
```

### 3. Buat program yang:
Mengisi stack dengan beberapa angka.
Menghitung jumlah elemen dalam stack tanpa menggunakan fungsi bawaan len()

```python
def kumpulkan():
    masuk = int(input("Masukkan Angka ke stack : "))
    angka = stack()
    push(angka,masuk)
    while True :
        pilihan = input("Lanjut? y/n")
        if pilihan == "y" :
            masuk = int(input("Masukkan Angka ke stack : "))
            push(angka,masuk)
        elif pilihan == "n" :
            return angka
            break
        else :
            print("pilihan anda tidak valid")
     
```

```python
def hitung(a):
    jumlah = 0
    while not isEmpty(a) :
        pop(a)
        jumlah += 1
    return jumlah
```

```python
angka = kumpulkan()
print('Jumlah Element : ',hitung(angka))
```

### 4. Diketahui stack berisi angka:

[10, 20, 30, 40, 50]

Buat program untuk menghapus angka 30 tanpa merusak urutan elemen lainnya.

```python
def menghapus(a, angka):
    hasil = stack()
    while True:
        if isEmpty(a):
            break
        wadah = pop(a)
        if wadah == angka:
            pass 
        else:
            push(hasil, wadah) 

    while True:
        if isEmpty(hasil):
            break
        else:
            push(a, pop(hasil)) 
            
    return a
```

```python
angka = [10,20,30,40,50]
hapus = 30
menghapus(angka,hapus)
```

### 6. Perpustakaan menyusun buku dalam bentuk stack.

Fitur program:

Tambah buku

Ambil buku paling atas

Tampilkan buku paling atas

Tampilkan seluruh buku

```python
def perpustakaan():
    rak = stack()
    while True :
        pilihan = int(input("Perpustakaan buku \n1. Tambah buku\n2. Ambil buku paling atas \n3. tampilkan buku paling atas \n4. tampilkan seluruh buku\n5. keluar \nMasukkan pilihan anda : "))
        if pilihan == 1 :
            buku = input("Masukkan Nama Buku : ")
            push(rak, buku)
            print(f'buku {buku} berhasil ditambahkan')
            print()
        elif pilihan == 2 :
            if isEmpty(rak) :
                print("Rak kosong!!!")
                print()
            else : 
                print(f'buku {pop(rak)} berhasil diambil')
                print()
        elif pilihan == 3 :
            if isEmpty(rak) :
                print("rak kosong!!!")
                print()
            else :
                print(f'buku paling atas di rak adalah : {peek(rak)}')
                print()
            print()
        elif pilihan == 4 :
            if isEmpty(rak) :
                print("rak kosong!!!")
                print()
            else :
                print(f'buku di rak saat ini {rak}')
                print()
        elif pilihan == 5 :
            break
        else :
            print('mohon pilih menu yang ada di daftar!!')
            print()
```
