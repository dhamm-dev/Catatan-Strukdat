# Definisi :
## 1. Counting Sort
Counting Sort adalah algoritma pengurutan yang sangat efisien karena tidak membandingkan elemen satu sama lain. Algoritma ini bekerja dengan cara menghitung jumlah kemunculan (frekuensi) dari setiap nilai unik di dalam daftar, lalu menggunakan informasi tersebut untuk menempatkan elemen tepat pada posisi akhirnya.

### Cara Kerja:
* Mencari nilai tertinggi (maksimum) di dalam daftar untuk menentukan rentang angka.
* Membuat daftar sementara (*count array*) untuk menghitung seberapa sering setiap angka muncul.
* Menjumlahkan nilai di *count array* secara kumulatif. Nilai kumulatif ini akan menjadi penunjuk indeks pasti di mana angka tersebut harus diletakkan.
* Menelusuri data awal dari belakang ke depan, lalu menempatkan setiap angka ke daftar hasil (*output array*) berdasarkan indeks kumulatifnya.
* Setiap kali angka ditempatkan, nilai kumulatifnya dikurangi satu.

# **CODE**

```python
# Fungsi Counting Sort
def counting_sort(arr):
    # Jika list kosong, langsung kembalikan
    if len(arr) == 0:
        return arr
        
    # 1. Cari nilai maksimum untuk menentukan ukuran array penghitung
    max_val = max(arr)
    
    # 2. Siapkan array penghitung (count) dan array hasil (output) dengan nilai awal 0
    count = [0] * (max_val + 1)
    output = [0] * len(arr)
    
    # 3. Hitung jumlah kemunculan setiap angka
    for num in arr:
        count[num] += 1
        
    # 4. Ubah array penghitung menjadi jumlah kumulatif
    for i in range(1, len(count)):
        count[i] += count[i - 1]
        
    # 5. Tempatkan angka ke posisi yang benar di array hasil 
    # (Ditelusuri dari belakang agar stabil)
    for i in range(len(arr) - 1, -1, -1):
        num = arr[i]
        output[count[num] - 1] = num
        count[num] -= 1
        
    return output

# ==========================================
# CONTOH PENGGUNAAN
# ==========================================
if __name__ == "__main__":
    data_acak = [54, 26, 93, 17, 1, 31, 2, 55, 20, 93, 20, 3, 4, 5]
    
    print("Data Awal             :", data_acak)
    
    hasil = counting_sort(data_acak)
    
    print("Data Setelah Diurutkan:", hasil)
```
