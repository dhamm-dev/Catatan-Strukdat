# Definisi :
## 1. Merge Sort
Merge Sort merupakan algoritma pengurutan dengan konsep rekursif yang terdiri dari dua tahapan utama, yaitu Splitting (divide) dan Merging. Algoritma ini didasarkan pada prinsip bahwa jika hanya terdapat satu elemen dalam sebuah list data, maka data tersebut otomatis sudah dalam keadaan terurut.

### Cara Kerja:
* **Tahap Divide (Splitting)**: Merupakan proses *Top-Down* di mana list data dibagi menjadi dua buah bagian secara terus menerus.
* Pembagian data ini dilakukan sampai hanya tersisa satu elemen saja didalam list.
* **Tahap Merging**: Merupakan tahapan *Bottom-up process* yang menggabungkan data-data yang sudah terbagi pada tahap *divide*.
* Proses merge ini melibatkan proses membandingkan data antara list yang akan digabung.
* Jika data lebih kecil, maka data tersebut akan diletakkan pada indeks yang lebih rendah.
* Proses ini akan dilakukan terus menerus, sehingga tidak ada lagi list yang akan digabungkan dan data pada list akan terurut sempurna.

# **CODE**

```python
def merge_sort(a_list):
    print("Splitting", a_list)
    
    # Akan terus dibagi selama panjang list lebih dari 1
    if len(a_list) > 1:
        mid = len(a_list) // 2
        left_half = a_list[:mid]
        right_half = a_list[mid:]
        
        # Pemanggilan rekursif untuk membagi sisi kiri dan kanan
        merge_sort(left_half)
        merge_sort(right_half)
        
        i = 0 # indeks untuk left_half
        j = 0 # indeks untuk right_half
        k = 0 # indeks hasil merging/sorting pada a_list
        
        # Proses Merging: Membandingkan elemen dari left_half dan right_half
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                a_list[k] = left_half[i]
                i = i + 1
            else:
                a_list[k] = right_half[j]
                j = j + 1
            k = k + 1
            
        # Memasukkan sisa elemen dari left_half (jika ada)
        while i < len(left_half):
            a_list[k] = left_half[i]
            i = i + 1
            k = k + 1
            
        # Memasukkan sisa elemen dari right_half (jika ada)
        while j < len(right_half):
            a_list[k] = right_half[j]
            j = j + 1
            k = k + 1
            
    print("Merging", a_list)


# ==========================================
# CONTOH PENGGUNAAN
# ==========================================
if __name__ == "__main__":
    a = [38, 12, 65, 22, 78, 25, 83, 21, 90]
    print("Data Awal:", a)
    print("-" * 30)
    
    merge_sort(a)
    
    print("-" * 30)
    print("Data Setelah Diurutkan:", a)
```
