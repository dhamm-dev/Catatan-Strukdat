# Definisi :
## 1. Quick Sort
Quick Sort adalah algoritma pengurutan yang tahapannya terdiri dari proses *divide* dan *combine*. Berbeda dengan algoritma *merge sort* yang membagi data menjadi dua bagian tanpa syarat, pembagian data pada *quick sort* harus memenuhi ketentuan berdasarkan nilai *pivot*.

### Cara Kerja:
* **Tahap Divide**: Membagi data menjadi dua buah bagian berdasarkan nilai *pivot*.
* Data yang berada di bagian sebelah kiri (bagian pertama) haruslah data-data yang memiliki nilai lebih kecil dari nilai *pivot*, meskipun urutannya masih acak.
* Data yang berada di sebelah kanan (bagian kedua) adalah data-data yang memiliki nilai lebih besar dari nilai *pivot*.
* Pada awal partisi, nilai *pivot* umumnya ditentukan dari data yang terdapat pada indeks awal dari sekumpulan data tersebut.
* Dilakukan penelusuran data; jika ditemukan data yang nilainya kurang dari *pivot*, maka data tersebut ditukar posisinya dengan data di sebelah *pivot*.
* Pada akhir partisi, *pivot* diletakkan di posisi indeks sebenarnya seolah-olah data tersebut sudah terurut.
* **Tahap Combine**: Tahapan ini pada dasarnya hanya memerintahkan proses partisi secara berulang kali pada sub-bagian data.
* Bagian data sebelum nilai *pivot* dan setelah nilai *pivot* akan dipartisi lagi menjadi dua bagian secara terus-menerus.
* Proses ini berlanjut sampai tidak ada lagi data yang bisa dipartisi, sehingga hasil akhirnya seluruh data akan terurut.

## **Divide**

Fungsi ini bertugas murni untuk menentukan pivot dan membagi data ke sisi kiri (lebih kecil) dan sisi kanan (lebih besar).

```python
# TAHAP 1: DIVIDE (Fungsi Partisi)
def partition(A, p, q):
    x = A[p]  # Menentukan pivot dari elemen pertama (indeks p)
    i = p     # Menyimpan batas elemen yang lebih kecil dari pivot
    
    # Menelusuri sisa data dari p+1 hingga q
    for j in range(p + 1, q + 1):
        # Jika elemen saat ini lebih kecil atau sama dengan pivot
        if A[j] <= x:
            i = i + 1
            # Tukar elemen agar berpindah ke sisi kiri
            A[i], A[j] = A[j], A[i]
            
    # Tukar posisi pivot ke tengah-tengah batas
    A[p], A[i] = A[i], A[p]
    print(A)
    # Kembalikan indeks posisi pivot yang baru
    return i
```

```python
a=[37,83,10,1,45,25,12,90,54]
partition(a,0,len(a)-1)
print(a)
```

## **Combine**

Fungsi ini bertugas untuk mengulang-ulang proses partisi di atas pada setiap potongan data sampai semuanya terurut sempurna. (Catatan: fungsi ini membutuhkan fungsi partition di atas agar bisa bekerja).

```python
# TAHAP 2: COMBINE (Fungsi Rekursif Quicksort)
def Quicksort(A, p, q):
    # Base case: Lakukan selama indeks awal (p) lebih kecil dari indeks akhir (q)
    if p < q:
        # Panggil tahap DIVIDE untuk memecah data dan mendapatkan indeks pivot (r)
        r = partition(A, p, q)
        
        # Panggil tahap COMBINE untuk mengurutkan bagian kiri pivot
        Quicksort(A, p, r - 1)
        
        # Panggil tahap COMBINE untuk mengurutkan bagian kanan pivot
        Quicksort(A, r + 1, q)
```
